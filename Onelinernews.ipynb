```
import os
```

```
base_dir = '/content/BBC News Summary/News Articles'
classes = ['business', 'sport', 'entertainment', 'tech', 'politics']
Summary=[]
i=0
for cl in classes:
  base_path = os.path.join(base_dir, cl)
  extention = '.txt'
  num = 1
  name = ''
  if num/100>1:
    name = '' + str(num)
  elif num/10>1:
    name = '0' + str(num)
  else:
    name= '00'+str(num)
  
  file_path=os.path.join(base_path, name)
  file_path= file_path + extention
  while True:
    if num/100>=1:
      name = '' + str(num)
    elif num/10>=1:
      name = '0' + str(num)
    else:
      name= '00'+str(num)
    file_path=os.path.join(base_path, name)
    file_path= file_path+extention
    try:
      f = open(file_path , "rb")
      print(name)
      bline=f.read()
      try:
        line=bline.decode("utf-8")
        Summary.append(line)
      except:
        pass
    except FileNotFoundError:
      break
    num=num+1
```
