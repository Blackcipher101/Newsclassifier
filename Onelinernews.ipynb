```
import os
```

```
base_dir = '/content/BBC News Summary/News Articles'
classes = ['business', 'sport', 'entertainment', 'tech', 'politics']
News=[[],[]]
labels=[]

num = 1
while num<511:
  extention = '.txt'
  
  name = ''
  if num/100>1:
    name = '' + str(num)
  elif num/10>1:
    name = '0' + str(num)
  else:
    name= '00'+str(num)
  j=0
  for cl in classes:
    base_path = os.path.join(base_dir, cl)
    file_path=os.path.join(base_path, name)
    file_path= file_path + extention
    if num/100>=1:
      name = '' + str(num)
    elif num/10>=1:
      name = '0' + str(num)
    else:
      name= '00'+str(num)
    num=num+1
    if num == 200: 
      continue
    print(num)
    file_path=os.path.join(base_path, name)
    file_path= file_path+extention
    try:
      f = open(file_path , "r")
      line=f.read()
      News[0].append(line)
      News[1].append(j)
      j=j+1
      
    except FileNotFoundError:
      pass
  
  
      
```

```
print(News[0][1],News[1][1])
```

```
def preprocess(News):
  News=News.replace('\n',' ')
  News=News.replace('\\','')
  return News
```

```
proNews=[]
for news in News[0]:
  proNews.append(preprocess(news))
```

```
proNews[0]
```

```
vocab_size = 500
embedding_dim = 24
max_length = 50
trunc_type='post'
padding_type='post'
oov_tok = "<OOV>"


from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

import tensorflow_datasets as tfds

vocab_size = 1000
tokenizer = tfds.deprecated.text.SubwordTextEncoder.build_from_corpus(proNews, vocab_size, max_subword_length=7)
```

```
print(proNews[4])
encoded = tokenizer.encode(proNews[4])
print(encoded)
```

```
for i, proNew in enumerate(proNews):
  proNews[i] = tokenizer.encode(proNew)
```

```
print(proNews[1])
```

```
max_length = 500
trunc_type='post'
padding_type='post'

# Pad all sentences
pronews_padded = pad_sequences(proNews, maxlen=max_length, 
                                 padding=padding_type, truncating=trunc_type)
```

```
training_size = int(len(proNews) * 0.8)
```

```
labels=News[1]
```

```
training_news = pronews_padded[0:training_size]
testing_news = pronews_padded[training_size:]
training_labels = labels[0:training_size]
testing_labels = labels[training_size:]
```

```
import numpy as np
training_labels_final = np.array(training_labels)
testing_labels_final = np.array(testing_labels)
```

```
training_labels_final[0],training_news[0]
```

```
num_epochs = 30

model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, embedding_dim, input_length=max_length),
    tf.keras.layers.GlobalAveragePooling1D(), 
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(5, activation='softmax')
])

learning_rate = 0.03

model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),optimizer=tf.keras.optimizers.Adam(learning_rate),metrics=['accuracy'])
model.summary()
```

```
epochs = 30
history = model.fit(training_news, training_labels_final, epochs=num_epochs, validation_data=(testing_news, testing_labels_final))
```

```
import matplotlib.pyplot as plt
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```
